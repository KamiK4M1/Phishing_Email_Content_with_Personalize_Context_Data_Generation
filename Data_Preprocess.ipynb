```python
import pandas as pd

# Helper function to extract and process data
def process_dataframe(df, email_col, body_col, label_col):
    # Normalize column names
    df = df.copy()

    # Filter phishing emails
    phishing_df = df

    # Combine email and body text
    if body_col in phishing_df.columns:
        phishing_df['combined_text'] = phishing_df[email_col].astype(str) + ' ' + phishing_df[body_col].astype(str)
    else:
        phishing_df['combined_text'] = phishing_df[email_col].astype(str)

    return phishing_df[['combined_text', label_col]]

# Combine all processed dataframes
all_dfs = [
    process_dataframe(df, 'Email Text', 'Email Type', 'Email Type'),
    process_dataframe(df1, 'subject', 'Body', 'label'),
    process_dataframe(df2, 'subject', 'Body', 'label'),
    process_dataframe(df3, 'subject', 'Body', 'label'),
    process_dataframe(df4, 'subject', 'Body', 'label'),
    process_dataframe(df5, 'subject', 'Body', 'label'),
    process_dataframe(df6, 'subject', 'Body', 'label')
]

# Concatenate into one dataframe
combined_df = pd.concat(all_dfs, ignore_index=True)

# Display the combined dataframe
print(combined_df)
```

```python
combined_df['label'] = combined_df['label'].replace(1, 'Phishing Email')
combined_df['label'] = combined_df['label'].replace(0, 'Safe Email')

# Step 2: Fill the "Email Type" column with values from the "label" column (if missing)
combined_df['Email Type'] = combined_df['Email Type'].fillna(combined_df['label'])

# Step 3: Drop the now redundant "label" column
combined_df = combined_df.drop(columns=['label'])

# Optional: Reset index for cleanliness
combined_df = combined_df.reset_index(drop=True)
```

```python
import pandas as pd

# Helper function to extract and process data
def process_dataframe(df, email_col, body_col, label_col):
    # Normalize column names
    df = df.copy()

    # Filter phishing emails
    phishing_df = df[df[label_col].isin([1, "Phishing Email"])]

    # Combine email and body text
    if body_col in phishing_df.columns:
        phishing_df['combined_text'] = phishing_df[email_col].astype(str) + ' ' + phishing_df[body_col].astype(str)
    else:
        phishing_df['combined_text'] = phishing_df[email_col].astype(str)

    return phishing_df[['combined_text', label_col]]

# Combine all processed dataframes
all_dfs = [
    process_dataframe(df, 'Email Text', 'Email Type', 'Email Type'),
    process_dataframe(df1, 'subject', 'Body', 'label'),
    process_dataframe(df2, 'subject', 'Body', 'label'),
    process_dataframe(df3, 'subject', 'Body', 'label'),
    process_dataframe(df4, 'subject', 'Body', 'label'),
    process_dataframe(df5, 'subject', 'Body', 'label'),
    process_dataframe(df6, 'subject', 'Body', 'label')
]

# Concatenate into one dataframe
combined_df = pd.concat(all_dfs, ignore_index=True)

# Display the combined dataframe
print(combined_df)
```

```python
combined_df['label'] = combined_df['label'].replace(1, 'Phishing Email')
combined_df['label'] = combined_df['label'].replace(0, 'Safe Email')

# Step 2: Fill the "Email Type" column with values from the "label" column (if missing)
combined_df['Email Type'] = combined_df['Email Type'].fillna(combined_df['label'])

# Step 3: Drop the now redundant "label" column
combined_df = combined_df.drop(columns=['label'])

# Optional: Reset index for cleanliness
combined_df = combined_df.reset_index(drop=True)
```

# Make it to JSON for finetune LLM

```python
import pandas as pd
import json

df = pd.read_csv("./Dataset/Finetune_Dataset.csv")

with open("./Dataset/finetune.jsonl", "w", encoding="utf-8") as f:
    for _, row in df.iterrows():
        email_type = row["Email Type"].strip().lower()

        if "phishing" in email_type:
            system_prompt = "You are a helpful assistant that writes realistic phishing emails for internal cybersecurity training."
            user_prompt = "Write a phishing email."
        else:
            system_prompt = "You are a helpful assistant that writes safe, professional emails."
            user_prompt = "Write a safe email."

        json_obj = {
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
                {"role": "assistant", "content": row["combined_text"].strip()}
            ]
        }
        f.write(json.dumps(json_obj, ensure_ascii=False) + "\n")
```
